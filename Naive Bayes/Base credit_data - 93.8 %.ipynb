#### Carregando a base de dados

```python
import pickle

with open(r'../.Material/Bases de dados/Bases pre-processadas/base_credit_data.pkl', mode='rb') as arquivo:
    x_previsores_treinamento, y_classes_treinamento, x_previsores_teste, y_classes_teste = pickle.load( arquivo )
```

#### Aplicação do método de aprendizagem bayeseana

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score  # Biblioteca para verificar acurácia da previsão do modelo

n_bayes = GaussianNB()
```

```python
n_bayes = n_bayes.fit( x_previsores_treinamento, y_classes_treinamento )
```

```python
previsao = n_bayes.predict( x_previsores_teste )
```

```python
from sklearn.metrics import accuracy_score

print('Acurácia da previsão: ', accuracy_score( y_classes_teste, previsao ) * 100, '%')
```

### Visualizando resultados do treinamento

```python
from yellowbrick.classifier import ConfusionMatrix, classification_report
from matplotlib.pyplot import xlabel, ylabel
```

#### Explicação do resultado

Nesta matriz, os valores são provenientes do cruzamento entre as previsões e os valores reais do conjunto de dados. Sendo assim, pode-se chegar à seguinte conclusão:

**Para a classe "0", 428 previsões estão corretas e 8 estão incorretas.**
<br>
**Para a classe "1", 23 previsões estão corretas e 41 estão incorretas."**


```python
c = ConfusionMatrix( n_bayes )
c.fit( x_previsores_treinamento, y_classes_treinamento )
c.score(x_previsores_teste, y_classes_teste)


xlabel('Previsões');
ylabel('Classes');
```

##### Gerando um relatório de classificação

```python
classification_report(
    n_bayes,  # Modelo do aprendizado de máquina
    x_previsores_treinamento,  # Previsores do treinamento
    y_classes_treinamento,  # Classes do treinamento
    x_previsores_teste,  # Previsores de teste
    y_classes_teste  # Classes de teste
);
```
